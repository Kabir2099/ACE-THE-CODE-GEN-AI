```python
from dotenv import load_dotenv
load_dotenv()
import os
groq_api_key = os.getenv('groq_api_key')
groq_api_key[:9]
from langchain_groq import ChatGroq
llm = ChatGroq(
    api_key = groq_api_key,
    model = 'llama-3.1-8b-instant'
)
```

# Conversation Memory

```python
from langchain_classic.chains.conversation.memory import ConversationBufferMemory
```

```python
memory = ConversationBufferMemory()
```

```python
from langchain_classic.chains import ConversationChain
```

```python
conversation = ConversationChain(

    llm = llm,
    memory = memory,
    verbose = True
    
)
```

```python
res = conversation.invoke('what is ML')
```

```python
res
```

```python
print(res['response'])
```

```python
res = conversation.invoke('5 benefits')
```

```python
print(res['history'])
```

```python
print(memory.buffer)
```

# Entity Memory

```python
from langchain_classic.chains.conversation.memory import ConversationEntityMemory
```

```python
from langchain_classic.chains.conversation.prompt import ENTITY_MEMORY_CONVERSATION_TEMPLATE
```

```python
print(ENTITY_MEMORY_CONVERSATION_TEMPLATE.template)
```

```python
memory = ConversationEntityMemory(llm = llm)
```

```python
chain = ConversationChain(
    llm = llm,
    memory = memory,
    prompt = ENTITY_MEMORY_CONVERSATION_TEMPLATE
)
```

```python
res = chain.invoke('Hi, I am kabir sk . I am from kolkata. I love to play football.')
```

```python
res['response']
```

```python
print(memory.buffer)
```
