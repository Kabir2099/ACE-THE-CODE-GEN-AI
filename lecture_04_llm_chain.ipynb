```python
from dotenv import load_dotenv
load_dotenv()
import os
groq_api_key = os.getenv('groq_api_key')
groq_api_key[:9]
from langchain_groq import ChatGroq
llm = ChatGroq(
    api_key = groq_api_key,
    model = 'llama-3.1-8b-instant'
)
```

# Simple Chain

```python
template = """
 ROLE:
You are an expert AI Travel Planner who specializes in creating budget-friendly, highly optimized, and experience-rich travel itineraries. You have deep knowledge of local attractions, hidden gems, food spots, transport, and cultural experiences.

TASK:
Create a detailed 7-day travel itinerary for the given city.

INPUT:
City: {city_name}

CONTEXT:
- The trip is for a traveler/couple who wants a balanced mix of sightseeing, local experiences, food exploration, and relaxation.
- Assume moderate budget unless specified.
- Include both popular tourist attractions and lesser-known hidden gems.
- Consider travel time between places and group nearby attractions logically.
- Include local food recommendations and unique experiences.

FORMAT:
Provide the response in the following structured format:

Day 1:
- Morning: [Place + activity]
- Afternoon: [Place + activity]
- Evening: [Place + activity]
- Food: [Local food recommendation + place]
- Travel Tips: [Transport / timing tip]

Day 2:
...

(Continue till Day 7)

At the end, include:
1. Total Estimated Budget (approx)
2. Best Areas to Stay
3. Transport Suggestions (local travel)
4. Must-Try Foods
5. Things to Avoid / Tips

CONSTRAINTS:
- Keep each day realistic (avoid too many places in one day)
- Optimize route (nearby places together)
- Avoid generic suggestions; be specific
- Include at least 1 unique/local experience per day
- Keep response clear, structured, and easy to read

  """
```

```python
from langchain_core.prompts import  PromptTemplate
```

```python
prompt = PromptTemplate.from_template(
    template = template,
    input_variable = ['city_name']
)
```

```python
llm_chain = prompt | llm 
```

```python
print(llm_chain.invoke('kolkata').content)
```

# chain with two variables

```python
template = """
ROLE:
You are an elite academic strategist and productivity coach who specializes in helping students crack exams efficiently using smart study techniques, spaced repetition, and active recall.

TASK:
Generate a highly optimized, personalized 7-day study plan for the given subject and available study hours.

INPUT:
Subject: {subject_name}
Hours per day: {hours}

CONTEXT:
- The student is preparing for exams and wants maximum output in limited time.
- The plan must balance concept learning, practice, and revision.
- Assume the student has average prior knowledge unless specified.
- Use scientifically proven techniques like active recall, spaced repetition, and interleaving.
- Avoid burnout by including smart breaks and light days.

THINKING STRATEGY:
- First, break the subject into logical subtopics (beginner → advanced).
- Then distribute them across 7 days with increasing difficulty.
- Allocate time per task based on importance and complexity.
- Ensure revision is spaced across multiple days.
- Include practice after every major topic.

FORMAT:
Return output in clean structured format:

Day 1:
- Focus Topics: [Specific topics]
- Study Plan (Hour-wise breakdown):
    - Hour 1: [Concept learning]
    - Hour 2: [Practice]
    - Hour 3: [Revision / Active Recall]
- Practice Tasks: [Type of questions or exercises]
- Revision Strategy: [What to revise from previous day]
- Break Plan: [Pomodoro / rest timing]
- Productivity Tip: [1 actionable tip]

(Repeat till Day 7)

FINAL OUTPUT SECTION:
1. 🔥 Key High-Weight Topics (most important for exam)
2. ⚠️ Weak Areas Strategy (how to improve fast)
3. 🔁 Smart Revision Plan (after 7 days)
4. 🎯 Exam Day Strategy
5. ⏱ Time Optimization Tips

CONSTRAINTS:
- Keep daily workload achievable within given hours
- Avoid vague instructions (be specific)
- Ensure progressive difficulty (easy → medium → hard)
- Include at least one revision loop every 2 days
- Do not overload Day 1 or Day 7
- Make plan actionable, not theoretical

OUTPUT QUALITY RULES:
- Be specific (mention actual topics, not “study basics”)
- Avoid repetition across days
- Prioritize clarity and usability
- Make it feel like a real mentor created it
"""
```

```python
prompt = PromptTemplate.from_template(
    template = template,
    input_variable = ['subject_name', 'hours']
)
```

```python
llm_chain = prompt | llm
```

```python
print(llm_chain.invoke({'subject_name': 'Data Structures', 'hours': 4}).content)
```

# Sequential Chain

```python
template_book_list = '''

You are an expert literary critic and librarian specializing in generating highly curated "Top 10" lists of books based on a specified genre. Your task is to analyze the user's genre input and produce a structured, high-quality list.
Genre: **{topic}**

**Constraints & Rules:**
1.  **Format:** The final output MUST be bulleted list.
2.  **Strictly adhere to the "Top 10" format.** Do not provide more or fewer than 10 entries.
3.  **Required Information:** For every book, you must provide:
    * **Title**
    * **Author**
    * A concise, one-sentence **Summary** that justifies its place in the top 10 for that genre.
4.  **Selection Criteria:** The list must include books that represent the genre's defining works, historical milestones, and most influential titles.


 '''
```

```python
prompt = PromptTemplate(

    template=template_book_list,
    input_variables=['topic']
)
```

```python
from langchain_classic.chains import LLMChain
```

```python
book_find_chain = LLMChain(
    llm=llm,
    prompt=prompt,
    output_key="book_list"
)
```

```python
result = book_find_chain.invoke({"topic": "self help"})
```

```python
print(result['book_list'])
```

```python
template_book_selection = """
You are an expert literary analyst. Your task is to process a list of 10 recommended books and select the SINGLE best book based on the following criteria:
1.  **Relevance:** How well does the book align with the specified genre?
2.  **Impact:** Consider the book's influence on literature and culture within the genre.
3.  **Critical Acclaim:** Take into account reviews, awards, and recognition the book has received.
 Analyze the.
**Constraints & Rules:**
1.  **Output Focus:** The response must only focus on the chosen book.
2.  **Required Sections:** The final output must be formatted with the three required sections: Chosen Book Details, Selection Rationale, and Brief Summary.
3.  **Tone:** Maintain a critical, insightful, and enthusiastic tone.

Analyze the following list of 10 books:
---
{book_list}
---


**FINAL OUTPUT FORMAT:**

## 🏆 Chosen Book Details
* **Title:** [Insert Title]
* **Author:** [Insert Author]
* **Original Publication Year:** [Insert Year]
* **Genre Tags:** [Insert 3-5 relevant sub-genres or themes]


## 📖 Brief Summary
[Provide a concise, spoiler-free paragraph (4-5 sentences) summarizing the book's core premise, primary conflict, and why it is a classic of the genre.]

 """
```

```python
book_summary_prompt = PromptTemplate(
    template=template_book_selection,
    input_variables=['book_list']
)
```

```python
book_summary_chain = LLMChain(llm=llm, prompt=book_summary_prompt, output_key='book_summary')
```

```python
res = book_summary_chain.invoke({'book_list': result['book_list']})
```

```python
print(res['book_summary'])
```

# sequential Chain

```python
from langchain_classic.chains import SequentialChain
```

```python
final_chain = SequentialChain(
    chains=[book_find_chain, book_summary_chain],
    input_variables=['topic'],
    output_variables=['book_list', 'book_summary'])
```

```python
res = final_chain.invoke({'topic': 'AI'})
```

```python
res
```

# Questions generation from text

```python
# generate summary from user input text
template = """ 
You are an expert editor and summarizer specializing in extracting core information from complex texts. Your sole task is to generate a concise, accurate, and objective summary of the user-provided text.

**Constraints:**
1.  **Strictly adhere to the requested summary length.**
2.  **Maintain an objective, neutral tone.** Do not add personal opinions or critical judgments.
3.  **Focus on the Main Points:** The summary must cover the central argument, key evidence, and final conclusion presented in the source text.

Please generate a summary of the following text:
---
**{user_input}**
---

**Final Output:** Present the summary directly without any introductory phrases or boilerplate text.
"""
```

```python
summary_prompt = PromptTemplate(
    template=template,
    input_variables=['user_input']
)
```

```python
summary_chain = LLMChain(llm=llm, prompt=summary_prompt, output_key='summary')
```

```python
template = """
You are a professional assessment designer and quiz creator. Your task is to analyze the provided book summary and generate a set of comprehensive, application-based questions that test understanding of the text's core concepts, plot points, and themes.
**Constraints & Rules:**
1.  **Question Types:** Include a mix of multiple-choice, short answer, and essay questions.
2.  **Depth of Understanding:** Questions should probe beyond surface-level details, encouraging critical thinking and analysis.

Input : {summary}

do not provide any explanations, only provide the questions."""
```

```python
question_prompt = PromptTemplate(
    template=template,
    input_variables=['summary'])
```

```python
question_chain = LLMChain(llm=llm, prompt=question_prompt, output_key='questions')
```

```python
final_chain = SequentialChain(
    chains=[summary_chain, question_chain],
    input_variables=['user_input'],
    output_variables=['summary', 'questions'])
```

```python
text = """Artificial Intelligence (AI) refers to the simulation of human intelligence in machines that are programmed to think and learn like humans. The field of AI encompasses a variety of subfields, including machine learning, natural language processing, computer vision, and robotics. AI systems can analyze large amounts of data, recognize patterns, and make decisions with minimal human intervention."""
```

```python
res = final_chain.invoke({'user_input':text})
```
